```python
import transformers
import torch
import os
```

```python
torch.set_num_threads(4)
```

```python
model_id = "../../Models/LLAMA3 HF/Meta-Llama-3-8B"
```

```python
os.environ["OPENMP_NUM_THREADS"] = "5"
```

```python
pipeline = transformers.pipeline(
    "text-generation", 
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
    max_new_tokens=1024

)
```

```python
pipeline("Hey how are you doing today?")
```
